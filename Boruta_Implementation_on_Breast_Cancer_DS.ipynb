```python
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

# Load the Breast Cancer dataset

```python
data = load_breast_cancer()
x = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

# Train-Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Define the Random Forest Model

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=1)
```

# Initialize Boruta

```python
boruta_selector = BorutaPy(rf, n_estimators='auto', verbose=2, random_state=42)
```

# Fit boruta on training data

```python
boruta_selector.fit(x_train.values, y_train)
```

# Print Selected Features

```python
selected_features = x_train.columns[boruta_selector.support_]
print("Selected Features:", selected_features)
```

# Print Tentative Features

```python
tentative_features = x_train.columns[boruta_selector.support_weak_]
print("Tentative Features:", tentative_features)
```

# Create a DataFrame for better visualization

```python
feature_importance_df = pd.DataFrame({
    'Feature': data.feature_names,
    'Ranking': boruta_selector.ranking_
})
feature_importance_df.sort_values(by='Ranking', ascending=True, inplace=True)
feature_importance_df = feature_importance_df.set_index('Feature')
feature_importance_df
```

# Display the DataFrame sorted by ranking

```python
print(feature_importance_df.sort_values(by='Ranking'))
```
